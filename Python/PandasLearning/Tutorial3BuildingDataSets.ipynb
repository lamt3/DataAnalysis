# Combining, Concatenating, and Appending DataFrames

```python
df1 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]},
                   index = [2001, 2002, 2003, 2004])

df2 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]},
                   index = [2005, 2006, 2007, 2008])

df3 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'Low_tier_HPI':[50, 52, 50, 53]},
                   index = [2001, 2002, 2003, 2004])
```

```python
#How to concatenate different data frames
concatDF = pd.concat([df1,df2,df3])
```

```python
#As you can see... it adds to the bottom of the list 2001,2002,etc instead of placing the values intuitively into
#above section in NaN for LowTierHPI
concatDF
```

```python
#Pandas not really meant to append data. More so meant for data manipulation based on existing data
df4 = df1.append(df2)
```

```python
df4
```

```python
df5 = df1.append(df3)
```

```python
df5
```

```python
#Better way to append data to existing data frame is to convert data into series and append it to corresponding dataframe
appendingSeries = pd.Series([80, 2, 50], index = ['HPI', 'Int_rate', 'US_GDP_Thousands'])
```

```python
df6 = df1.append(appendingSeries, ignore_index = True)
```

# Joining and Merging Dataframes

```python
df1 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]},
                   index = [2001, 2002, 2003, 2004])

df2 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]},
                   index = [2005, 2006, 2007, 2008])

df3 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Unemployment':[7, 8, 9, 6],
                    'Low_tier_HPI':[50, 52, 50, 53]},
                   index = [2001, 2002, 2003, 2004])
```

```python
#Creates Duplicate Data
pd.merge(df1,df2, on='HPI')
```

```python
#To fix you can merge on multiple columns
pd.merge(df1,df2, on=['HPI', 'Int_rate'])
```

```python
#Joining
#Now share an index but no columns for df1 and df3
df1.set_index('HPI', inplace=True)
df3.set_index('HPI', inplace=True)
joined = df1.join(df3)
```

```python
joined
```

```python
#What if 2 different data frames don't share any indexes?
df1 = pd.DataFrame({'Year':[2001, 2002, 2003, 2004],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]})


df3 = pd.DataFrame({'Year':[2001, 2003, 2004, 2005],
                    'Unemployment':[7, 8, 9, 6],
                    'Low_tier_HPI':[50, 52, 50, 53]})
merged = pd.merge(df1,df3, on='Year')
merged.set_index('Year', inplace=True)
```

```python
#Notice how 2002 and 2005 data is missing
merged
```

```python
#Merges have 4 types of Merges
#1. Left
    # pd.merge('left', 'right', on='variable', how='left' )
#2. Right
#3. Outer
    #Union of the keys --> will display all data on the joined column 
#4. Inner
    #The default merge --> where the keys intersect 
```

```python
#Left Merge Example
mergedLeft = pd.merge(df1, df3, on = 'Year', how='left')
mergedLeft.set_index('Year', inplace=True)
```

```python
#Prints all the values from df1 index... see NaN for Low Tier HPI and Unemployment in 2002
#b/c there is no corresponding data which is expected
mergedLeft
```

```python
#Outer Merge Example
mergedOuter = pd.merge(df1, df3, on = 'Year', how='outer')
mergedOuter.set_index('Year', inplace=True)
```

```python
#Prints each value from 2001-2005
mergedOuter
```

```python
#Use JOIN when index matters
#Use MERGE when index does NOT matter
```
